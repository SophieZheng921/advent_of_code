```python
input_data=pl.read_csv('/Users/lihong.zheng/Projects/advent_of_code/day_1/data.csv')
input_data

```

```python


df_sorted_multi = input_data.select(pl.col(['List_1', ' List_2']).sort())
df_sorted_multi 
```

```python


sub=df_sorted_multi.select(pl.col("List_1")-pl.col(" List_2"))
sub
```

```python

lst = sub.get_column('List_1').to_list()
lst
```

```python
sum=0
for a in lst:
    if a<0:
        sum+=(-a)
    else:
        sum+=a

print(sum)
```
